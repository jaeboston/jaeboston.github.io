```python
import pandas as pd
```

```python
df = pd.read_excel("S:/Research/TrendStrategies/PriceDatabase/EQUITY_ETF_SCR.xlsx",sheet_name='Sheet1')
```

```python
df.columns
```

```python
ID2keep = ['XLC','XLY','XLP','XLE','XLF','XLV','XLI','XLB','XLRE','XLK','XLU']
```

```python
df.head()
```

```python
sData_COVERAGE_TCS           = df.pivot(index='DATE',columns='ID',values='COVERAGE_TCS')
sData_COVERAGE_TCS_ADJ       = df.pivot(index='DATE',columns='ID',values='COVERAGE_TCS_ADJ')
sData_COVERAGE_PCF_REL       = df.pivot(index='DATE',columns='ID',values='COVERAGE_PCF_REL')
sData_COVERAGE_PCF_ADJ_REL   = df.pivot(index='DATE',columns='ID',values='COVERAGE_PCF_ADJ_REL')

sData_TCS_Final              = df.pivot(index='DATE',columns='ID',values='TCS_Final')
sData_PCF_Final              = df.pivot(index='DATE',columns='ID',values='PCF_Final')
sData_TCS_ADJ_Final          = df.pivot(index='DATE',columns='ID',values='TCS_ADJ_Final')
sData_PCF_ADJ_Final          = df.pivot(index='DATE',columns='ID',values='PCF_ADJ_Final')
```

```python
sData_COVERAGE_TCS.head()
```

```python
#: keep the columns in ID2keep
sData_COVERAGE_TCS          = sData_COVERAGE_TCS[ID2keep]
sData_COVERAGE_TCS_ADJ      = sData_COVERAGE_TCS_ADJ[ID2keep]
sData_COVERAGE_PCF_REL      = sData_COVERAGE_PCF_REL[ID2keep]
sData_COVERAGE_PCF_ADJ_REL  = sData_COVERAGE_PCF_ADJ_REL[ID2keep]

sData_TCS_Final             = sData_TCS_Final[ID2keep]
sData_PCF_Final             = sData_PCF_Final[ID2keep]
sData_TCS_ADJ_Final         = sData_TCS_ADJ_Final[ID2keep]
sData_PCF_ADJ_Final         = sData_PCF_ADJ_Final[ID2keep]
```

```python
sData_COVERAGE_TCS.head()
```

## Using Dictionary to create multi label columns

```python

dd = {}

dd['COVERAGE_TCS']         = sData_COVERAGE_TCS
dd['COVERAGE_TCS_ADJ']     = sData_COVERAGE_TCS_ADJ
dd['COVERAGE_PCF_REL']     = sData_COVERAGE_PCF_REL
dd['COVERAGE_PCF_ADJ_REL'] = sData_COVERAGE_PCF_ADJ_REL

dd['TCS_Final']            = sData_TCS_Final
dd['PCF_Final']            = sData_PCF_Final
dd['TCS_ADJ_Final']        = sData_TCS_ADJ_Final
dd['PCF_ADJ_Final']        = sData_PCF_ADJ_Final

df_T = pd.concat(dd, axis=1)
```

```python
df_T.head()
```
